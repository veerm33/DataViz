```python
import pandas as pd
import matplotlib. pyplot as plt
import numpy as np
```

```python
df = pd.read_csv(r'/Users/HappyPlace/Desktop/2012_nba_draft_combine.csv')
df.head()
```

```python
#For my peers and colleagues, I want them to compare players by thier agility and sprint speed in order to determine the overall fastest player in this years draft combine.
```

```python
df_sorted = df.sort_values(['Agility', 'Sprint'], ascending=[True, True])


df_top15 = df_sorted[['Player', 'Agility', 'Sprint']].head(15)


fig, ax = plt.subplots(figsize=(8, 6))
ax.barh(df_top15['Player'], df_top15['Agility'], color='blue', label='Agility')
ax.barh(df_top15['Player'], df_top15['Sprint'], color='red', label='Sprint')
ax.set_xlabel('Time (seconds)')
ax.set_title('Top 15 Fastest Performers in Agility and Sprint')
ax.legend()
plt.show()
```

```python
#Here, peers will be able to view the top 15 fastest performers in the draft class. 
#Combined with the knowlege of player height and weight, they can reference the list to determine why Andre Drummond is a great pick for team in need of a big player.
#Andre Drummond is listed at 7 ft tall and weighs in at 279 lbs. He has a tremendous speed advantage compared to other players at his size and position. 
```

```python
top_10 = df.sort_values('Draft pick').head(10)
top_10 = top_10[['Player', 'Sprint', 'Agility']]
top_10 = top_10.set_index('Player')


ax = top_10.plot(kind='bar', figsize=(40,26), color=['blue', 'red'], rot=0)
plt.xlabel('Player (In order of draft pick selection)', fontsize=35)
ax.set_ylabel('Time (seconds)', fontsize=35)
ax.set_title('Top 10 NBA Draft Picks: Sprint and Agility Times', fontsize=35)


ax.tick_params(axis='x', labelsize=22)
ax.tick_params(axis='y', labelsize=26)

plt.legend(['Agility', 'Sprint'], fontsize=20)

ax.xaxis.labelpad = 30


plt.show()
```

```python
#This visual will show my peers the top ten draft picks and thier combined sprint and agility times.
#They will immediately understand that the first and fourth pick did not perform the speed and agility drills, which means their selections were based on previous player knowledge and other combine results. 
```

```python
import seaborn as sns
```

```python
cols = ['Player', 'Agility', 'Sprint']

df_heatmap = df[cols]


fig, ax = plt.subplots(figsize=(50,70))


df_heatmap.set_index('Player', inplace=True)


ax.set_xlabel('Skills', fontsize=30)
ax.set_ylabel('Players', fontsize=50)

ax.tick_params(axis='x', labelsize=50)
ax.tick_params(axis='y', labelsize=35)



sns.heatmap(df_heatmap, cmap='RdPu_r')

cbar = ax.figure.colorbar(ax.collections[0])
cbar.ax.tick_params(labelsize=30)

plt.show()
```

```python
#For this heatmap, I ran into the problem where higher values were represented by Darker colors.

#This is the opposite of what I needed.

#Since smaller numbers are representing faster speeds, I wanted those figures to stand out more.

#I sought computer science help from work colleagues and they showed me a few color sprectrums that would work.

#This variation of color wouldn't be helpful to the casual viewer, but for my peers who are studying this dataset in depth, this is a fast way of displaying the fastest players in the combine.
```
