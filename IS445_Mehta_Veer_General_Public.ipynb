```python
import pandas as pd
import matplotlib. pyplot as plt
import numpy as np
```

```python
df = pd.read_csv(r'/Users/HappyPlace/Desktop/2012_nba_draft_combine.csv')
df.head()
```

```python
df_top3 = df.sort_values('Draft pick').head(3)

fig, ax = plt.subplots(figsize=(30, 18))
df_top3.plot(kind='barh', x='Player', y=['Height (No Shoes)', 'Wingspan', 'Standing reach'], color=['blue', 'orange', 'green'], ax=ax)
plt.title('Top 3 Draft Picks - Height, Wingspan, and Reach', fontsize=50, pad=15)
plt.xlabel('Measurement (inches)', fontsize=30)
plt.ylabel('Player', fontsize=30)


ax.tick_params(axis='x', labelsize=30, pad=15)
ax.tick_params(axis='y', labelsize=30, pad=15)

plt.legend(['Height (No Shoes)', 'Wingspan', 'Standing reach'], fontsize=24)
plt.show()
```

```python
#This visual clearly shows the general audience measurements taken for the top 3 draft selections. 
```

```python
df_top3 = df[df['Draft pick'].isin([2, 3])]

fig, ax = plt.subplots(figsize=(30, 18))

df_top3.plot(kind='barh', x='Player', y=['Vertical (Max)', 'Vertical (Max Reach)', 'Vertical (No Step)', 'Vertical (No Step Reach)'], color=['pink', 'purple', 'orange', 'red'], ax=ax)

plt.title('Top 3 Draft Picks - Verticals', fontsize=50)
plt.xlabel('Measurement (inches)', fontsize=30)
plt.ylabel('Player', fontsize=30)


ax.tick_params(axis='x', labelsize=30, pad=10)
ax.tick_params(axis='y', labelsize=30, pad=10)

plt.legend(['Max Vertical', 'Vertical with Reach', 'Standing Vertical Leap', 'Standing Leap with Reach'], fontsize=24)
plt.show()
```

```python

# Get the top 20 draft picks
top_20 = df.head(20)

# Divide the top 20 draft picks based on height (with shoes) above or below 78 inches
above_78 = top_20[top_20['Height (With Shoes)'] > 78]
below_78 = top_20[top_20['Height (With Shoes)'] <= 78]

# Create a bar chart
heights = ['Above 78 inches', 'Below or equal to 78 inches']
counts = [len(above_78), len(below_78)]
plt.bar(heights, counts)

# Set the chart title and axis labels
plt.title('Top 20 Draft Picks Categorized by Height')
plt.xlabel('Height (with shoes)')
plt.ylabel('Number of draft picks')

# Display the chart
plt.show()
```

```python
#This visual shows the general public the correclation of NBA teams selecting taller players over shorter ones. 
```
