```python
#This dataset provides the 2012 NBA draft combine report and measurements.
#I wanted to pick a dataset for a topic that I aim to seek in the job market:Sports statistics 
#This dataset lists player names along with their height, wingspan, reach, body fat%, weight, vertical, Handsize, bench press result, sprint result, and agility test result. 

# https://data.world/achou/nba-draft-combine-measurements
```

```python
import pandas as pd
import matplotlib. pyplot as plt
import numpy as np
```

```python
df = pd.read_csv(r'/Users/HappyPlace/Desktop/2012_nba_draft_combine.csv')
df.head()
```

```python
height_by_player = df.groupby('Player')['Height (No Shoes)'].sum()
```

```python
height_by_player.plot(kind='bar')

fig, ax = plt.subplots(figsize=(30, 20))

height_by_player.plot(kind='bar', ax=ax)

plt.xlabel('Player', fontsize=35)
plt.ylabel('Height (No Shoes)', fontsize=35)

ax.set_title('Height (No Shoes) by Player', fontsize=40)

ax.tick_params(axis='x', labelsize=22)
ax.tick_params(axis='y', labelsize=25)

# Plot the data as a bar graph with wider bars
height_by_player.plot(kind='bar', ax=ax, width=0.4, edgecolor='black', linewidth=1.0)

# Set the y-axis limits to be between 60 and 80
ax.set_ylim(60, 90)

plt.show()

#This graph gives me the general data of players by their height. I can quickly descern the tallest players from the shortest players.
```

```python
vertical_by_player = df.groupby('Player')['Vertical (Max)'].sum()
```

```python
vertical_by_player.plot(kind='bar')

fig, ax = plt.subplots(figsize=(30, 20))

vertical_by_player.plot(kind='bar', ax=ax)

plt.xlabel('Player', fontsize=35)
plt.ylabel('Vertical', fontsize=35)

ax.set_title('Maximum Vertical by Player', fontsize=40)

ax.tick_params(axis='x', labelsize=22)
ax.tick_params(axis='y', labelsize=25)

height_by_player.plot(kind='bar', ax=ax, width=0.4, edgecolor='black', linewidth=1.0)

ax.set_ylim(65, 85)

plt.show()

#This is the same parameters as the graph above but lists players by their maximum vertical. 
#I can use these two lists to single out the players that have the best vertical to height ratio.
```

```python
BodyFat_by_player = df.groupby('Player')['Body Fat'].sum()
```

```python
BodyFat_by_player.plot(kind='bar')

fig, ax = plt.subplots(figsize=(30, 20))

BodyFat_by_player.plot(kind='bar', ax=ax)

plt.xlabel('Player', fontsize=35)
plt.ylabel('Body Fat %', fontsize=35)

ax.set_title('Body Fat Percentage by Player', fontsize=40)

ax.tick_params(axis='x', labelsize=22)
ax.tick_params(axis='y', labelsize=25)

BodyFat_by_player.plot(kind='bar', ax=ax, width=0.4, edgecolor='black', linewidth=1.0)


plt.show()

#With the two graphs above, I can use this visual to determine which players have more room for vertical improvement if they cut some percentage of body fat.
```
